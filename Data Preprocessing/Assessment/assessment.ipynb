### 7. Create a line plot for x and y

```python
import matplotlib.pyplot as plt

x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Line Plot")
plt.show()
```

### 8. Plot x, x², x³, x⁴ with styles and axis limits

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6.1, 0.3)

plt.plot(x, x, "r")     # x
plt.plot(x, x**2, "b")  # x²
plt.plot(x, x**3, "g")   # x³
plt.plot(x, x**4, "c")   # x⁴

plt.xlim(0, 6)
plt.ylim(0, 125)
plt.xlabel("x")
plt.ylabel("values")
plt.title("Powers of x")
plt.show()
```

### 9. Bar plot of heights with names

```python
import matplotlib.pyplot as plt

height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

plt.bar(names, height)
plt.title("Height Comparison")
plt.xlabel("Names")
plt.ylabel("Height (cm)")
plt.show()
```

### 10. Histogram of 100,000 random normal points

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100000)

# 10 bins
plt.hist(x, bins=10)
plt.title("Histogram (10 bins)")
plt.show()

# 20 bins
plt.hist(x, bins=20)
plt.title("Histogram (20 bins)")
plt.show()

# 50 bins
plt.hist(x, bins=50)
plt.title("Histogram (50 bins)")
plt.show()
```
